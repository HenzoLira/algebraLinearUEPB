# 03 — Espaços Vetoriais
**Referência:** Boldrini et al., Cap. 4, p. 97–135

---

## 3.1 Motivação

A ideia de espaço vetorial **generaliza** os vetores geométricos: qualquer conjunto onde possamos somar elementos e multiplicá-los por escalares, respeitando 8 propriedades naturais, é um espaço vetorial. Isso permite tratar matrizes, polinômios e funções com as **mesmas ferramentas**.

---

## 3.2 Definição (Boldrini, p. 103)

Um **espaço vetorial real** é um conjunto $V$ com duas operações:
- **Adição:** $u + v \in V$ para quaisquer $u, v \in V$
- **Multiplicação por escalar:** $\alpha \cdot v \in V$ para qualquer $\alpha \in \mathbb{R}$, $v \in V$

---

## 3.3 Os 8 Axiomas

Para quaisquer $u, v, w \in V$ e $\alpha, \beta \in \mathbb{R}$:

| # | Axioma | Nome |
|---|--------|------|
| (i) | $u + v = v + u$ | Comutatividade |
| (ii) | $(u+v)+w = u+(v+w)$ | Associatividade |
| (iii) | $\exists\, \mathbf{0} \in V : u + \mathbf{0} = u$ | Vetor nulo |
| (iv) | $\exists\, {-u} \in V : u + (-u) = \mathbf{0}$ | Oposto |
| (v) | $\alpha(u+v) = \alpha u + \alpha v$ | Distributividade I |
| (vi) | $(\alpha+\beta)u = \alpha u + \beta u$ | Distributividade II |
| (vii) | $(\alpha\beta)u = \alpha(\beta u)$ | Associatividade escalar |
| (viii) | $1 \cdot u = u$ | Neutro escalar |

---

## 3.4 Exemplos de Espaços Vetoriais

- $\mathbb{R}^n$ — n-uplas de reais (Boldrini, p. 104)
- $M(m,n)$ — matrizes reais $m \times n$ (Boldrini, p. 104)
- $P_n$ — polinômios de grau $\leq n$ (Boldrini, p. 105)


```python
from sympy import Matrix, symbols, zeros
from sympy import init_printing
init_printing(use_unicode=True)

# ── Verificando axiomas em M(2,2) simbolicamente ──────
a, b, c, d, e, f, g, h = symbols('a b c d e f g h')
lam, mu = symbols('lambda mu')

U = Matrix([[a, b], [c, d]])
V = Matrix([[e, f], [g, h]])
W = Matrix([[1, 0], [0, 1]])  # exemplo concreto
O = zeros(2, 2)

print("Verificando os 8 axiomas no espaço M(2,2):")
print()
print(f"(i)  U + V == V + U?        {U + V == V + U}")
print(f"(ii) (U+V)+W == U+(V+W)?    {(U+V)+W == U+(V+W)}")
print(f"(iii) U + O == U?            {U + O == U}")
print(f"(iv)  U + (-U) == O?         {U + (-U) == O}")
print(f"(v)   λ(U+V) == λU + λV?    {lam*(U+V) == lam*U + lam*V}")
print(f"(vi)  (λ+μ)U == λU + μU?    {(lam+mu)*U == lam*U + mu*U}")
print(f"(vii) (λμ)U == λ(μU)?       {(lam*mu)*U == lam*(mu*U)}")
print(f"(viii) 1·U == U?             {1*U == U}")
print()
print("✅ Todos os axiomas verificados — M(2,2) é espaço vetorial!")
```

```python
# ── Exemplos de espaços vetoriais ────────────────────
import numpy as np

print("Exemplo 1: R^n")
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])
print(f"  u = {u}")
print(f"  v = {v}")
print(f"  u + v = {u+v}")
print(f"  3·u = {3*u}")
print(f"  Vetor nulo: {np.zeros(3)}")

print("\nExemplo 2: P_2  (polinômios de grau ≤ 2)")
print("  p(x) = 1 + 2x        q(x) = 3 - x + x²")
print("  p+q  = 4 + x + x²    (soma de polinômios)")
print("  2·p  = 2 + 4x         (escalar)")
print("  Vetor nulo: p(x) = 0")
```

```python
# ── Visualização: R^2 como espaço vetorial ────────────
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

u = np.array([2, 1])
v = np.array([1, 3])

# Plot 1: soma vetorial
ax = axes[0]
origin = np.array([0, 0])
ax.quiver(*origin, *u, color='royalblue', angles='xy', scale_units='xy', scale=1, width=0.015, label=f'u = {u}')
ax.quiver(*origin, *v, color='tomato', angles='xy', scale_units='xy', scale=1, width=0.015, label=f'v = {v}')
ax.quiver(*origin, *(u+v), color='green', angles='xy', scale_units='xy', scale=1, width=0.015, label=f'u+v = {u+v}')
ax.quiver(*u, *v, color='tomato', angles='xy', scale_units='xy', scale=1, width=0.008, alpha=0.4, linestyle='dashed')
ax.quiver(*v, *u, color='royalblue', angles='xy', scale_units='xy', scale=1, width=0.008, alpha=0.4, linestyle='dashed')
ax.set_xlim(-1, 5); ax.set_ylim(-1, 5)
ax.axhline(0, color='gray', lw=0.5); ax.axvline(0, color='gray', lw=0.5)
ax.grid(True, alpha=0.3); ax.legend(); ax.set_title('Axioma (i): u+v = v+u', fontsize=12)

# Plot 2: multiplicação escalar
ax = axes[1]
for alpha, color in [(0.5, 'lightblue'), (1, 'royalblue'), (2, 'darkblue'), (-1, 'tomato')]:
    av = alpha * u
    ax.quiver(*origin, *av, color=color, angles='xy', scale_units='xy', scale=1, width=0.015, label=f'α={alpha}, αu={av}')
ax.set_xlim(-3, 5); ax.set_ylim(-2, 3)
ax.axhline(0, color='gray', lw=0.5); ax.axvline(0, color='gray', lw=0.5)
ax.grid(True, alpha=0.3); ax.legend(fontsize=9); ax.set_title('Multiplicação por escalar', fontsize=12)

plt.suptitle('ℝ² como Espaço Vetorial', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
