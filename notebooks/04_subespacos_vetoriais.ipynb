# 04 — Subespaços Vetoriais
**Referência:** Boldrini et al., Seção 4.3, p. 105–112

---

## 4.1 Motivação

Dentro de $\mathbb{R}^3$, o plano $xy$ (onde $z=0$) tem uma propriedade especial: somar dois vetores do plano dá um vetor do plano; multiplicar por escalar também. Esse plano é um **subespaço** de $\mathbb{R}^3$.

Um subespaço é um subconjunto que **funciona como espaço vetorial por conta própria**.

---

## 4.2 Definição (Boldrini, p. 106)

Um subconjunto **não vazio** $W \subseteq V$ é um **subespaço vetorial** de $V$ se:

$$\text{(i)}\; \forall\, u, v \in W \Rightarrow u+v \in W \quad \text{(fechado na soma)}$$

$$\text{(ii)}\; \forall\, \alpha \in \mathbb{R},\, u \in W \Rightarrow \alpha u \in W \quad \text{(fechado no escalar)}$$

> 💡 **Consequência imediata:** todo subespaço contém o **vetor nulo** (tome $\alpha = 0$ em (ii)). Logo, todo subespaço **passa pela origem**.

---

## 4.3 Subespaços Triviais

Todo espaço $V$ tem pelo menos dois subespaços triviais: $\{\mathbf{0}\}$ e o próprio $V$.

---

## 4.4 Como Verificar

Para provar que $W$ é subespaço, basta verificar as **2 condições** — os outros 6 axiomas se herdam automaticamente de $V$.


```python
from sympy import Matrix, symbols
from sympy import init_printing
init_printing(use_unicode=True)

x2, x3, y2, y3, alpha = symbols('x2 x3 y2 y3 alpha')

# ── Exemplo 1: W = {(0, x2, x3)} em R³ ──────────────
print("W = {(0, x2, x3) ∈ R³}  (1ª coordenada sempre 0)")
print()

u = Matrix([0, x2, x3])
v = Matrix([0, y2, y3])
soma = u + v
escalar = alpha * u

print(f"(i)  u + v = {soma.T}")
print(f"     1ª coordenada: {soma[0]} = 0  ✓")
print()
print(f"(ii) α·u = {escalar.T}")
print(f"     1ª coordenada: {escalar[0]} = 0  ✓")
print()
print("✅ W é subespaço de R³")
```

```python
# ── Exemplo 2 (contraexemplo): W = {x+y=1} em R² ─────
print("W = {(x,y) ∈ R² : x + y = 1}  (NÃO é subespaço)")
print()

# Dois vetores em W
u = Matrix([1, 0])  # 1+0=1 ✓
v = Matrix([0, 1])  # 0+1=1 ✓
soma = u + v

print(f"u = {u.T}  (1+0=1 ✓)")
print(f"v = {v.T}  (0+1=1 ✓)")
print(f"u + v = {soma.T}  → 1+1={soma[0]+soma[1]} ≠ 1  ✗")
print()
print("Falha na condição (i) → NÃO é subespaço!")
print()
print("Confirmação: (0,0) ∉ W pois 0+0=0≠1 → não passa pela origem")
```

```python
# ── Visualização: subespaços de R² e R³ ──────────────
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(15, 5))

# --- Plot 1: Subespaço em R² (reta pela origem) ---
ax1 = fig.add_subplot(131)
t = np.linspace(-3, 3, 100)
v_dir = np.array([1, 2])
ax1.plot(t * v_dir[0], t * v_dir[1], 'royalblue', lw=2, label='W = [(1,2)] ← subespaço')
ax1.plot(0, 0, 'ko', ms=8, label='Origem ∈ W')
ax1.set_xlim(-3, 3); ax1.set_ylim(-3, 3)
ax1.axhline(0, color='gray', lw=0.5); ax1.axvline(0, color='gray', lw=0.5)
ax1.grid(True, alpha=0.3); ax1.legend(fontsize=8)
ax1.set_title('Subespaço: reta pela origem em ℝ²', fontsize=10)

# --- Plot 2: NÃO subespaço (reta fora da origem) ---
ax2 = fig.add_subplot(132)
t = np.linspace(-3, 3, 100)
ax2.plot(t, 1 - t, 'tomato', lw=2, label='x+y=1 ← NÃO é subespaço')
ax2.plot(0, 0, 'ko', ms=8, label='Origem ∉ conjunto')
ax2.set_xlim(-3, 3); ax2.set_ylim(-3, 3)
ax2.axhline(0, color='gray', lw=0.5); ax2.axvline(0, color='gray', lw=0.5)
ax2.grid(True, alpha=0.3); ax2.legend(fontsize=8)
ax2.set_title('NÃO subespaço: reta fora da origem', fontsize=10)

# --- Plot 3: Plano em R³ ---
ax3 = fig.add_subplot(133, projection='3d')
xx, yy = np.meshgrid(np.linspace(-2, 2, 10), np.linspace(-2, 2, 10))
zz = np.zeros_like(xx)
ax3.plot_surface(xx, yy, zz, alpha=0.3, color='royalblue')
ax3.set_xlabel('x'); ax3.set_ylabel('y'); ax3.set_zlabel('z')
ax3.set_title('Subespaço: plano z=0 em ℝ³', fontsize=10)

plt.suptitle('Subespaços Vetoriais', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 📝 Exercícios

1. Verifique se $W = \{(x,y,z) \in \mathbb{R}^3 : x + y + z = 0\}$ é subespaço.
2. Verifique se $W = \{(x,y) \in \mathbb{R}^2 : xy = 0\}$ é subespaço.
3. É verdade que a interseção de dois subespaços é também um subespaço? Justifique.


```python
from sympy import symbols, Matrix

x1, x2, x3, y1, y2, y3, a = symbols('x1 x2 x3 y1 y2 y3 alpha')

# Ex. 1: x+y+z=0
u = Matrix([x1, x2, -x1-x2])
v = Matrix([y1, y2, -y1-y2])
soma = u + v
print("Ex.1: W = {x+y+z=0}")
print(f"  u+v = {soma.T}")
print(f"  Soma das coords: {soma[0]+soma[1]+soma[2]} = 0  ✓")
print(f"  α·u = {(a*u).T}, soma coords = {a*u[0]+a*u[1]+a*u[2]} = 0  ✓")
print("  ✅ W é subespaço")

# Ex. 2: xy=0 — contra-exemplo
print("\nEx.2: W = {xy=0} — contra-exemplo")
u2 = Matrix([1, 0])  # 1*0=0 ✓
v2 = Matrix([0, 1])  # 0*1=0 ✓
s2 = u2 + v2
print(f"  u=(1,0)∈W, v=(0,1)∈W, u+v={s2.T}: {s2[0]*s2[1]}≠0  ✗ NÃO é subespaço")
```
