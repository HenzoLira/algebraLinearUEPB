## 1.1 O que é uma Matriz?

Imagine que você precisa organizar as notas de quatro alunos em três provas. Uma forma natural é montar uma tabela: cada **linha** é um aluno, cada **coluna** é uma prova. Essa tabela retangular de números é exatamente o que chamamos de **matriz**.

### Definição formal (Boldrini, p. 2)

Uma **matriz $m \times n$** é uma tabela retangular de elementos organizada em $m$ linhas e $n$ colunas:

$$A = [a_{ij}], \quad i = 1,\ldots,m \;\; j = 1,\ldots,n$$

O elemento $a_{ij}$ está na $i$-ésima linha e $j$-ésima coluna.

> **Convenção:** letras maiúsculas para matrizes ($A$, $B$), letras minúsculas com dois índices para elementos ($a_{ij}$).

---

## 1.2 Tipos Especiais de Matrizes

| Tipo | Condição |
|------|----------|
| **Quadrada** | $m = n$ |
| **Nula** | $a_{ij} = 0$ para todo $i, j$ |
| **Diagonal** | Quadrada com zeros fora da diagonal principal |
| **Identidade** $I_n$ | Diagonal com $1$'s na diagonal |
| **Transposta** $A^T$ | $(A^T)_{ij} = a_{ji}$ |
| **Simétrica** | $A = A^T$ |

---

## 1.3 Operações com Matrizes

### Soma
Só entre matrizes de **mesma ordem**: $(A+B)_{ij} = a_{ij} + b_{ij}$

### Multiplicação por escalar
$(kA)_{ij} = k \cdot a_{ij}$

### Multiplicação de matrizes
Possível quando **colunas de $A$ = linhas de $B$**. Se $A$ é $m\times n$ e $B$ é $n\times p$, então $C = A \cdot B$ é $m\times p$:
$$c_{ij} = \sum_{k=1}^{n} a_{ik} \cdot b_{kj}$$

> ⚠️ **Atenção:** $AB \neq BA$ em geral — a multiplicação **não é comutativa**!


```python
from sympy import Matrix, eye, zeros, pprint, latex
from sympy import init_printing
init_printing(use_unicode=True)

print("=" * 50)
print("  MATRIZES — Exemplos com SymPy")
print("=" * 50)

# ── Definindo matrizes ────────────────────────────────
A = Matrix([[1, -2, 0],
            [4, -3, 2]])
print("\nMatriz A (2×3):")
pprint(A)
print(f"Ordem: {A.shape[0]}×{A.shape[1]}")
print(f"Elemento a_12 = {A[0, 1]}  (linha 1, coluna 2)")
print(f"Elemento a_23 = {A[1, 2]}  (linha 2, coluna 3)")
```

```python
# ── Tipos especiais ───────────────────────────────────
print("Identidade 3×3:")
pprint(eye(3))

print("\nMatriz Nula 2×3:")
pprint(zeros(2, 3))

B = Matrix([[1, 2], [3, 4]])
print("\nB:")
pprint(B)
print("\nTransposta B^T:")
pprint(B.T)
print(f"\nB é simétrica? {B == B.T}")

C = Matrix([[1, 0], [0, 5]])
print("\nC (diagonal):")
pprint(C)
print(f"C é simétrica? {C == C.T}")
```

```python
# ── Operações ─────────────────────────────────────────
print("── Soma ──")
A = Matrix([[3, 1], [0, 5]])
B = Matrix([[-1, 2], [4, 0]])
print("A ="); pprint(A)
print("B ="); pprint(B)
print("A + B ="); pprint(A + B)

print("\n── Multiplicação por escalar ──")
print("3 * A ="); pprint(3 * A)
```

```python
# ── Multiplicação de matrizes ─────────────────────────
print("── Multiplicação A · B ──")
C = Matrix([[1, 2], [3, 4]])
D = Matrix([[5, 6], [7, 8]])

print("C ="); pprint(C)
print("D ="); pprint(D)
print("\nC · D ="); pprint(C * D)
print("\nD · C =  (diferente!)"); pprint(D * C)
print(f"\nC·D == D·C? {C*D == D*C}  ← não comutativo!")
```

```python
# ── Visualização: não-comutatividade ──────────────────
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

C_np = np.array([[1, 2], [3, 4]], dtype=float)
D_np = np.array([[5, 6], [7, 8]], dtype=float)
CD = C_np @ D_np
DC = D_np @ C_np

for ax, M, title in zip(axes, [CD, DC], ['C · D', 'D · C']):
    im = ax.imshow(M, cmap='Blues', aspect='auto')
    ax.set_title(title, fontsize=14, fontweight='bold')
    for i in range(2):
        for j in range(2):
            ax.text(j, i, f'{M[i,j]:.0f}', ha='center', va='center',
                    fontsize=16, color='black')
    ax.set_xticks([])
    ax.set_yticks([])
    plt.colorbar(im, ax=ax)

plt.suptitle('AB ≠ BA  —  Multiplicação de Matrizes não é Comutativa', fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# ── Espaço para seus exercícios ───────────────────────
from sympy import Matrix

# Ex. 1
A = Matrix([[2, -1], [0, 3]])
print("A^T ="); pprint(A.T)
print(f"A é simétrica? {A == A.T}")

# Ex. 2
A2 = Matrix([[1, 0], [2, 1]])
B2 = Matrix([[3, 1], [-1, 2]])
print("\nA · B ="); pprint(A2 * B2)

# Ex. 3
from sympy import eye
I = eye(2)
print(f"\nI · A == A? {I * A2 == A2}")
```
